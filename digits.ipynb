```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train[0].shape
```

```python
X_train.shape
X_train
```

```python
X_test.shape
print(X_test)
```

```python
xtrain=X_train.reshape(len(X_train),28*28)
xtest=X_test.reshape(len(X_test),28*28)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(xtrain, y_train, epochs=5)
```

```python
from sklearn.preprocessing import StandardScaler
StandardScaler.fit_transform(xtrain)
```

```python
StandardScaler.fit_transform(xtest)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(xtrain, y_train, epochs=10)
```

```python
model.evaluate(xtest,y_test)
```

```python
pred=model.predict(xtest)
```

```python
plt.matshow(X_test[0])
```

```python
ylabels = [np.argmax(i) for i in pred]
```

```python
ylabels
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=ylabels)
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(xtrain, y_train, epochs=5)
```

```python
model.evaluate(xtest,y_test)
```
